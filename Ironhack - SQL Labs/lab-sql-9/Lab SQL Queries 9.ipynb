```python
data = pd.read_sql_query('SELECT title FROM sakila.film ORDER BY length desc LIMIT 10', engine)
data.index = data.index+1
data.index.names = ['Rank']
data.columns = ['10 Longest Movies']
data
```

```python
data = pd.read_sql_query("SELECT COUNT(special_features) FROM sakila.film WHERE special_features LIKE '%Behind the Scenes%'", engine)
data.columns = ['Movies having behind the scenes']
data
```

```python
data = pd.read_sql_query("SELECT rating, (AVG(length)) as Avg_Duration_in_mn FROM sakila.film GROUP BY rating HAVING Avg_Duration_in_mn>=120", engine)
data.columns = 'Categories', 'Average Duration in mn'
data
```

```python
data = pd.read_sql_query('SELECT title, length, CASE WHEN length <= 60 then "duration low" WHEN length <= 120 then "duration medium" ELSE "duration high" END AS "Ranking duration" FROM sakila.film WHERE length is not null AND length <> 0;', engine)
data.columns = 'Movie', 'Length(mn)', 'Ranking duration'
data
```
